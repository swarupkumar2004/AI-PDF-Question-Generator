```python
import os

# Rename for easier use
os.rename("IMO class 1 Maths Olympiad Sample Paper 1 for the year 2024-25.pdf", "IMO_Sample.pdf")
```

```python
import pdfplumber

def extract_text_from_pdf(pdf_path):
    extracted_text = []
    with pdfplumber.open(pdf_path) as pdf:
        for i, page in enumerate(pdf.pages, start=1):
            text = page.extract_text()
            extracted_text.append({
                "page": i,
                "text": text.strip() if text else ""
            })
    return extracted_text

text_data = extract_text_from_pdf("IMO_Sample.pdf")
```

```python
import fitz  # PyMuPDF
import os

def extract_images_from_pdf(pdf_path, output_folder="extracted_images"):
    os.makedirs(output_folder, exist_ok=True)
    doc = fitz.open(pdf_path)
    image_data = []

    for page_num in range(len(doc)):
        page = doc[page_num]
        images = page.get_images(full=True)
        for img_index, img in enumerate(images):
            xref = img[0]
            base_image = doc.extract_image(xref)
            image_bytes = base_image["image"]
            img_ext = base_image["ext"]
            image_filename = f"page{page_num+1}_image{img_index+1}.{img_ext}"
            image_path = os.path.join(output_folder, image_filename)

            with open(image_path, "wb") as f:
                f.write(image_bytes)

            image_data.append({
                "page": page_num + 1,
                "image_path": image_path
            })

    return image_data

image_data = extract_images_from_pdf("IMO_Sample.pdf")
```

```python
import json

def create_structured_json(text_data, image_data, output_file="output.json"):
    page_map = {}

    for item in text_data:
        page_map[item["page"]] = {
            "text": item["text"],
            "images": []
        }

    for img in image_data:
        if img["page"] not in page_map:
            page_map[img["page"]] = {
                "text": "",
                "images": []
            }
        page_map[img["page"]]["images"].append(img["image_path"])

    structured_output = []
    for page in sorted(page_map.keys()):
        structured_output.append({
            "page": page,
            "text": page_map[page]["text"],
            "images": page_map[page]["images"]
        })

    with open(output_file, "w", encoding="utf-8") as f:
        json.dump(structured_output, f, indent=2)

    print(f"✅ JSON saved as {output_file}")

create_structured_json(text_data, image_data)
```

```python
from transformers import BlipProcessor, BlipForConditionalGeneration
from PIL import Image
import torch

device = "cuda" if torch.cuda.is_available() else "cpu"

processor = BlipProcessor.from_pretrained("Salesforce/blip-image-captioning-base", use_fast=True)
model = BlipForConditionalGeneration.from_pretrained("Salesforce/blip-image-captioning-base").to(device)
```

```python
import os

def generate_captions(image_folder):
    captions = []

    for filename in sorted(os.listdir(image_folder)):
        if filename.endswith(".png") or filename.endswith(".jpg"):
            image_path = os.path.join(image_folder, filename)
            raw_image = Image.open(image_path).convert('RGB')

            inputs = processor(raw_image, return_tensors="pt").to(device)
            out = model.generate(**inputs)
            caption = processor.decode(out[0], skip_special_tokens=True)

            captions.append({
                "image": image_path,
                "caption": caption
            })

    return captions

# Run it
image_captions = generate_captions("extracted_images")
```

```python
def generate_question_from_caption(caption_text):
    return f"What is shown in the image: '{caption_text}'?"

questions = []
for entry in image_captions:
    question_text = generate_question_from_caption(entry['caption'])
    questions.append({
        "question": question_text,
        "image": entry['image'],
        "caption": entry['caption']
    })
```

```python
import json

with open("ai_generated_questions.json", "w") as f:
    json.dump(questions, f, indent=2)

print("✅ Saved as ai_generated_questions.json")
```

```python
bad_keywords = ["number of the numbers", "aminos", "the logo", "diagram of the"]

cleaned_questions = [q for q in questions if not any(bad in q["caption"].lower() for bad in bad_keywords)]

print(f"Filtered {len(questions) - len(cleaned_questions)} unclear questions.")
```

```python
def refine_question(caption):
    if "cat" in caption:
        return f"Which image shows {caption}?"
    elif "flower" in caption:
        return f"What is the flower shown in the image?"
    elif "number" in caption:
        return f"Which image represents the number {caption.split()[-1]}?"
    else:
        return f"What is shown in the image: '{caption}'?"

for q in questions:
    q["question"] = refine_question(q["caption"])
```

```python
import random

def add_mcq_options(data):
    for q in data:
        other_images = [x["image"] for x in data if x["image"] != q["image"]]
        q["options"] = random.sample(other_images, 3) + [q["image"]]
        random.shuffle(q["options"])
    return data

mcq_questions = add_mcq_options(cleaned_questions)
```

```python
import json
with open("final_image_questions.json", "w") as f:
    json.dump(mcq_questions, f, indent=2)

from google.colab import files
files.download("final_image_questions.json")
```
